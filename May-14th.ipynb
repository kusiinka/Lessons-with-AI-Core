```
#wave opens .WAV files
import wave
import numpy as np
import matplotlib.pyplot as plt
```

```
# open file into a wave object
signal_wave = wave.open('/content/gdrive/My Drive/data/audio.wav', 'r')
#get the sample rate
sample_rate = signal_wave.getframerate()
#reading the first 15 seconds because sample rate is samples per second, so * 15, it is 15 seconds worth of samples
# if parameter had been -1, it would have read the whole file
read_signal = signal_wave.readframes(sample_rate*15)
print(read_signal)
#loading what was read into a numpy array
sig = np.frombuffer(read_signal, dtype=np.int16)

print(sig.shape)
```

```
print(sig.shape)
#So if there are two channels, every alternating data point is a different channel
if signal_wave.getnchannels() == 2:
    left, right = sig[0::2], sig[1::2]
    #audio channels. 
    print(left.shape)
    print(right.shape)

# figure is a type of visualisation 
plt.figure()
#sig is the numpy array
plt.plot(sig)
plt.xlabel('sample rate * time')
plt.ylabel('energy')
plt.show()

plt.figure()
plt.specgram(sig, Fs=sample_rate)
plt.xlabel('Time')
plt.ylabel('Frequency')

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#new dataframe for the insurance data
df = pd.read_csv('/content/gdrive/My Drive/data/AutoInsurSweden.csv')
#print dataframes's head to check what the column headers are
print(df.head())
```

```
x = np.array(df['number_of_claims'])
y = np.array(df['total_payment(thousands of Swedish Kronor)'])
plt.scatter(x, y)
```

```
#challenge was to make a function x to predict y
# y= x is gradient of 1 
# this line is slightly off, so its a human guess
def predict(x):
    return 3.6*x
y_pred = predict(x)

plt.xlabel('number_of_claims')
plt.ylabel('total_payment(thousands of Swedish Kronor)')
plt.scatter(df['number_of_claims'], df['total_payment(thousands of Swedish Kronor)'])
plt.plot(x, y_pred, c='red')
```

```
#what are we doing here?
# sum ( (pred - y)^2 for all point) / lengthof(pred or y)
# y value is total payments

#cost is mean of all squared differences between real y points and predicted points
# (squared bcause we do not want negative value for cost)

def cost(y_pred, y):
    diff = y_pred-y
    c = np.mean(diff**2)
    return c
```

```
plt.xlabel('number_of_claims')
plt.ylabel('total_payment(thousands of Swedish Kronor)')
plt.scatter(df['number_of_claims'], df['total_payment(thousands of Swedish Kronor)'])
plt.plot(x, y_pred, c='red')
print(cost(y_pred, y))
```

```
# pupils analytical solution to finding a predictor of the predict function
# this is a closed form solution for a straight line fit .... so if the line is more complex, it won't work very well
def predict_predictor(x, y):
    y_mean = np.mean(y)
    x_mean = np.mean(x)
    grad = y_mean/x_mean
    print('Found gradient', grad)
    def predictor(a):
        return grad * a
    return predictor
```

```
predictor = predict_predictor(x, y)
y_pred = predictor(x)
plt.xlabel('number_of_claims')
plt.ylabel('total_payment(thousands of Swedish Kronor)')
plt.scatter(df['number_of_claims'], df['total_payment(thousands of Swedish Kronor)'])
plt.plot(x, y_pred, c='red')
print(cost(y_pred, y))
```
